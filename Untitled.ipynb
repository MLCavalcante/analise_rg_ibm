```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
rh = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

# Análise Univariada

```python
_ = plt.hist(rh.Age)
_ = plt.title('Histograma de idade')
```

```python
_ = plt.hist(rh.MonthlyIncome)
_ = plt.title('Histograma de Salário')
```

```python
rh.Gender.value_counts()
```

Como é uma empresa que contrata muitas pessoas da área de STEM, é esperado o número superior de homens.

```python
rh.EducationField.value_counts()
```

```python
_ = rh.EducationField.value_counts(1).plot.bar()
_ = plt.title('Headcount por Campo Educacional')
```

Vemos que a maioria dos funcionários é formado na área de Life Sciences. Na sequência temos a área Medical. A área com menos funcionáparios é Human Resources.  

#  Análise Bivariada

```python
_ = sns.boxplot(x='Gender', y ="MonthlyIncome", data=rh)
```

Curiosamente, a distribuição salarial entre homens e mulheres é parecida. O gender gap parece não ocorrer.

```python
_ = sns.boxplot(x='Attrition', y='Age', hue='Gender', data=rh)
```

Aparentemente, há uma maior proporção de mulheres que permaneceram na empresa. Mulheres mais jovens tendem a deixar a ampresa com mais frequência, mas ainda em menor número que homens.  

```python
_ = sns.boxplot(x='EducationField', y ="MonthlyIncome", data=rh)
```

O salário parece não ter uma relação forte com a área.  

```python
_ = sns.boxplot(x='PerformanceRating', y ="MonthlyIncome", data=rh)
```

Aparentemente o melhor desempenho se concentra entre os menores salários. 

```python
_ = sns.boxplot(x='Attrition', y='EnvironmentSatisfaction', data=rh)

```

Aparentemente o ambiente de trabalho não tem relação com a saída de funcionários. 

```python
_ = sns.scatterplot(x ='YearsAtCompany', y ='MonthlyIncome', data = rh)
```

Aparentemente o salário não tem relação com o tempo de empresa. 

# Conclusão 

Pode-se concluir que não há uma relação forte e direta entre a maioria das variáveis avaliadas nessa exploração. Isso não significa necessariamente que as variáveis são independentes ou que não existe relação alguma entre elas, mas apenas que a relação não é forte o suficiente para ser identificada por meio de correlação

Correlação não linear.
